# SubPlot
plt.subplot(n,m,i) or plt.subplot(nmi) 
n represent the row, m represent the column and i represent the position of the chart

```python
plt.figure(figsize=(6,4))
#plot 1
x=np.array([0,1,2,3])
y=np.array([3,8,1,10])
plt.subplot(1,3,1)
plt.plot(x,y)
plt.title('Plot 1')

#plot 2
b=np.array([0,1,2,3])
c=np.array([10,20,30,40])
plt.subplot(1,3,2)
plt.plot(b,c)
plt.title('Plot 2')

#plot 3
d=np.array([0,1,2,3])
e=np.array([10,20,30,40])
plt.subplot(1,3,3)
plt.plot(d,e)
plt.title('Plot 3')


plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,6))
left=fifa.loc[fifa['Preferred Foot']=='Left'].count(0)[0]
right=fifa.loc[fifa['Preferred Foot']=='Right'].count(0)[0]
labels=['left foot','right foot']
colors=['m','y']
explode=[.1,.3]
plt.title('Foot Preference')

plt.pie([left,right],labels=labels,colors=colors,autopct='%.3f%%',explode=explode)
plt.legend

plt.show
```
